```python
rows = 5

for i in range(1, rows + 1):
    for j in range(i):
        print("*", end="")
    print()
```

```python
rows = 5

for i in range(1, rows + 1):
    # Print spaces
    for j in range(rows - i):
        print(" ", end="")
    # Print stars
    for k in range(2 * i - 1):
        print("*", end="")
    print()
```

```python
rows = 5

for i in range(1, rows + 1):
    for j in range(1, i + 1):
        print(j, end="")
    print()
```

```python
rows = 5 # You can change this value to adjust the height of the pattern

for i in range(rows, 0, -1):
    for j in range(1, i + 1):
        print(j, end="")
    print()
```

```python
rows = 5 # You can change this value to adjust the size of the diamond

# Upper part of the diamond (pyramid)
for i in range(1, rows + 1):
    for j in range(rows - i):
        print(" ", end="")
    for k in range(2 * i - 1):
        print("*", end="")
    print()

# Lower part of the diamond (inverted pyramid)
for i in range(rows - 1, 0, -1):
    for j in range(rows - i):
        print(" ", end="")
    for k in range(2 * i - 1):
        print("*", end="")
    print()
```

```python
rows = 5 # You can change this value to adjust the size of the hollow diamond

# Upper part of the hollow diamond
for i in range(1, rows + 1):
    for j in range(rows - i):
        print(" ", end="")
    print("*", end="")
    if i > 1:
        for k in range(1, 2 * i - 2):
            print(" ", end="")
        print("*", end="")
    print()

# Lower part of the hollow diamond
for i in range(rows - 1, 0, -1):
    for j in range(rows - i):
        print(" ", end="")
    print("*", end="")
    if i > 1:
        for k in range(1, 2 * i - 2):
            print(" ", end="")
        print("*", end="")
    print()
```

```python
rows = 5 # You can change this value to adjust the number of rows in Pascal's triangle

for i in range(rows):
    num = 1
    print("  " * (rows - i - 1), end="") # Print leading spaces for centering
    for j in range(i + 1):
        print(f"{num:4d}", end="") # Print number with fixed width for alignment
        num = num * (i - j) // (j + 1)
    print()
```

```python
rows = 5 # You can change this value to adjust the number of rows in Floyd's triangle
current_number = 1

for i in range(1, rows + 1):
    for j in range(i):
        print(current_number, end=" ")
        current_number += 1
    print()
```

```python
rows = 5 # You can change this value to adjust the size of the butterfly

# Upper half of the butterfly
for i in range(1, rows + 1):
    # Left wing
    for j in range(i):
        print("*", end="")
    # Spaces between wings
    for j in range(2 * (rows - i)):
        print(" ", end="")
    # Right wing
    for j in range(i):
        print("*", end="")
    print()

# Lower half of the butterfly
for i in range(rows, 0, -1):
    # Left wing
    for j in range(i):
        print("*", end="")
    # Spaces between wings
    for j in range(2 * (rows - i)):
        print(" ", end="")
    # Right wing
    for j in range(i):
        print("*", end="")
    print()
```

```python
rows = 6 # You can adjust this value to change the size of the heart

# Upper part of the heart
for i in range(rows // 2, rows):
    # Left side of the heart
    for j in range(rows - i - 1):
        print(" ", end="")
    for j in range(2 * i - rows):
        print("*", end="")

    # Space between two upper bumps
    for j in range(2 * (rows - i - 1)):
        print(" ", end="")

    # Right side of the heart
    for j in range(2 * i - rows):
        print("*", end="")
    print()

# Lower part of the heart (inverted pyramid)
for i in range(rows, 0, -1):
    for j in range(rows - i):
        print(" ", end="")
    for j in range(2 * i - 1):
        print("*", end="")
    print()
```

```python
# Cat pattern program

def print_cat_pattern():
    print("  /\\_/\\  ")
    print(" ( o.o )  ")
    print("  > ^ <   ")
    print(" /  \\ /  \\")
    print("(___)(___)")

print_cat_pattern()
```

```python
# Messi pattern: Representing his jersey number '10'

def print_messi_pattern():
    print("  _______")
    print(" /  ___  \\ ")
    print("|  /   \\  |")
    print("|  \\_ _/  |   _______ ")
    print("|   ___   |  /  ___  \\ ")
    print("|  |   |  | |  /   \\  |")
    print("|  |___|  | |  \\_ _/  |")
    print(" \\_______/   \\_______/")

print_messi_pattern()
```

```python
# Generic Face Pattern

def print_face_pattern():
    print("   _____")
    print("  / o o \\")
    print(" |   ^   |")
    print(" |  \\_/  |")
    print("  \\_____/")

print_face_pattern()
```
